```python
dataset.apply(lambda x: sum(x.isnull()),axis=0) # checking missing values in each column of train dataset
```

```python
##create a list of columns where we can't have zeroes
#clean the data with NANs\#replace 0 with No data NaN
zero_not_accepted=['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area','Loan_Status']
for column in zero_not_accepted:
  #dataset[column]=dataset[column].replace(0,np.NaN)
  #mean=int(dataset[column].mean(skipna=True))
  dataset[column]=dataset[column].replace(np.NaN,0)
```
